```python
def relu (tensor):
    return max (0.0, tensor)
```

### Hyperbolic Tangent

```python
x
```

```python
from math import exp
```

```python
def tanh (tensor):
    return (exp (tensor) - exp (-tensor)) / (exp (tensor) + exp (-tensor))
```

```python
output = [tanh (i) for i in x]
```

```python
plt.plot (x, output)
```

### Sigmoid

```python
def sigmoid (z):
    return 1 / (1 + exp (-z))
```

```python
output = [sigmoid (i) for i in x]
```

```python
plt.plot (x, output)
```

#### Use seed to stabilize the random outcomes

```python
np.random.seed (2)
outputLayer = np.random.randn(1, 5)
```

```python
outputLayer
```

```python
def softmax (x):
    return np.exp (x) / np.exp (x).sum ()
```

```python
result = softmax (outputLayer)
```

```python
result
```

```python
np.argmax (result)
```

```python
x = np.arange (-10, 10, 1, dtype = float)
```

```python
x
```

```python
def leakyRelu (x, alpha = 0.05):
    return max (x * alpha, x)
```

```python
output = [leakyRelu (i) for i in x]
```

```python
plt.plot (x, output)
```
